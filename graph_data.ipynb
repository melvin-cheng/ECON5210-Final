```python
import matplotlib.pyplot as plt
from retrieve.get_new_data import get_yield
import numpy as np

def plot_yields(term, use_weekly=False, use_diff=False):
    df = get_yield(term=term)
    df = df.ffill()

    if use_weekly:
        df = df.resample("W-FRI").last()
    if use_diff:
        df = df.diff().dropna()

    plt.figure(figsize=(12, 6))

    colors = plt.cm.tab20(np.linspace(0, 1, len(df.columns)))

    for col, color in zip(df.columns, colors):
        plt.plot(df.index, df[col], label=col, color=color)

    plt.title(f"{term}-Year Yields")
    plt.xlabel("Date")
    plt.ylabel("Yield")
    plt.legend()
    plt.show()
```

# Daily Yield Plots

```python
plot_yields(1)
```

```python
plot_yields(2)
```

```python
plot_yields(3)
```

```python
plot_yields(5)
```
